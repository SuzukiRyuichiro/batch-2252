### Define a `capitalize_name` method which takes `first_name` and `last_name` as parameters and returns the well-formatted fullname (with capitalized first and last names).

```ruby
def capitalize_name(first_name, last_name)
  c_first_name = first_name.capitalize
  c_last_name = last_name.capitalize
  full_name = "#{c_first_name} #{c_last_name}"

  return full_name
end

puts capitalize_name('john', 'doe')
puts capitalize_name('leonardo', 'dicapprio')
```
